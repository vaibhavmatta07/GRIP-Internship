# Task - Prediction using Supervised ML
# Predict the percentage of a student based on number of study hours

```python
# Importing the libraries required

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print("Imported Successfully")
```

```python
%matplotlib inline
print("Done")
```

# Reading and Exploring the Data

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported")
```

```python
# Getting the first 10 values of dataset
data.head(10)
```

```python
# To determine number of rows and columns
data.shape
```

```python
# Brief of data
data.info()
```

```python
# Statistical Summary
data.describe()
```

# Data Plotting/Visualization

```python
sns.pairplot(data)
```

```python
data.plot(x='Hours',y='Scores', style='.')
plt.title("Hours Vs Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Scored")
plt.show()
```

```python
sns.heatmap(data.corr(),annot = True, cmap = "Accent")
```

```python
# x is input variable
x= data['Hours']
# y is output or dependent variable
y= data['Scores']
```

```python
x = np.array(x)
x = x.reshape(-1,1)
```

# Splitting the Data into Training and Test Sets

```python
from sklearn.model_selection import train_test_split
print('Successfully Imported')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=0)
```

# Training the Model

```python
from sklearn.linear_model import LinearRegression
lrm = LinearRegression()
lrm.fit(x_train, y_train)

print('Training Completed')
```

```python
# Getting the parameters i.e Intercept and Coefficient
print("Intercept is ",lrm.intercept_)
print("Coefficient is ", lrm.coef_)
```

```python
# Plotting the regression line for test data

line = lrm.coef_*x+lrm.intercept_  
plt.scatter(x, y)  
plt.plot(x, line);  
plt.show()
```

# Making Predictions

```python
print(x_test) # Testing data - In Hours
pred = lrm.predict(x_test) # Predicting the scores
```

```python
pred
```

# Comparing Actual vs Predicted

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': pred})
df
```

```python
# Plotting Actual vs Predicted

df.plot(kind='bar', figsize=(5.5,5.5))
plt.xlabel('Hours')
plt.ylabel('Percentage Scored')
plt.show()
```

# Evaluating the Model

```python
Accuracy = lrm.score(x_test,y_test)
print("Accuracy of Model is ", Accuracy*100, "%")
```

```python
from sklearn import metrics 
print('Mean Absolute Error :{}'.format(metrics.mean_absolute_error(y_test, pred)))
print('Mean Squared Error :{}'.format(metrics.mean_squared_error(y_test, pred)))
print('Root Mean Square Error :{}'.format(np.sqrt(metrics.mean_squared_error(y_test, pred))))
print('r-2:', metrics.r2_score(y_test,pred))
```

# Testing against given number of hours

```python
hours = 9.25
test_pred = lrm.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(test_pred[0]))
```
